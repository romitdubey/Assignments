Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Matplotlib is a Python library used for data visualization. It provides a wide range of tools for creating static, animated, and interactive plots in Python.

Matplotlib is widely used because it is flexible, powerful, and customizable. It allows users to create a wide variety of charts and plots including line plots, scatter plots, bar plots, histograms, and more.

Here are five common plots that can be created using the Pyplot module of Matplotlib:

1. Line Plot
2. Bar Plot
3. Scatter Plot
4. Histogram
5. Pie chart

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
#A scatter plot is used to visualize the relationship between two variables. It is created using the plt.scatter() function.

from matplotlib import pyplot as plt

plt.scatter(x,y)
plt.xlabel('x Axis')
plt.ylabel('y Axis')
plt.title('Random Data')
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

Use the following data:
    


```python
import numpy as np
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
from matplotlib import pyplot as plt

#plot1
plt.subplot(2,2,1)
plt.plot(x1,y1)

#plot2
plt.subplot(2,2,2)
plt.plot(x2,y2)

#plot3
plt.subplot(2,2,3)
plt.plot(x3,y3)

#plot4
plt.subplot(2,2,4)
plt.plot(x4,y4)

plt.suptitle("Example of Subplot")
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

A bar plot is used to show the comparison between different categories. It is created using the plt.bar() function.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

from matplotlib import pyplot as plt

plt.subplot(2,1,1)
plt.bar(company,profit)
plt.title('Bar')

plt.subplot(2,1,2)
plt.barh(company,profit)
plt.title('Horizontal Bar')

plt.tight_layout()
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

A box plot, also known as a box-and-whisker plot, is a type of chart used to display the distribution of a dataset. It provides a visual summary of the data, including the median, quartiles, and outliers.

```python
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
from matplotlib import pyplot as plt

plt.subplot(1,2,1)
plt.boxplot(box1)

plt.subplot(1,2,2)
plt.boxplot(box2)
plt.show()
```
