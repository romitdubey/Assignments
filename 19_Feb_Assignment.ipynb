Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a micro web framework for building web applications in Python. It is designed to be lightweight and modular, with minimal dependencies and a simple, easy-to-use API. Flask allows developers to quickly and easily create web applications with features such as routing, templating, and database integration.

###Some advantages of using Flask include:

Lightweight and modular: Flask is a minimalistic framework, which means it does not have a lot of built-in features. However, this also means that it is lightweight, easy to use, and can be customized to meet the needs of a particular project.

Flexibility: Flask is designed to be flexible and can be used to build a wide variety of web applications, from simple one-page apps to complex, multi-page applications.

Easy to learn: Flask is known for its simple and intuitive API, making it easy for developers to learn and use. Its small size and minimalistic design also mean that developers do not need to learn a lot of new concepts or technologies to use it effectively.

Extensible: Flask can be extended with a wide range of third-party libraries and plugins, making it a versatile and powerful tool for building web applications.

Large community: Flask has a large and active community of developers who contribute to the framework's development, documentation, and support. This means that developers can benefit from a wealth of resources and expertise when working with Flask.

Q3. What is App routing in Flask? Why do we use app routes?

App routing in Flask refers to the process of mapping URLs to specific functions or views within a Flask application. In other words, app routing determines which code should be executed when a particular URL is requested by a client.

We use app routes in Flask to create different pages or views within a web application. For example, we can define a route to display the homepage of an application, another route to handle user authentication, and yet another route to display a specific item or resource.

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.

```python
#In Flask, the url_for() function is used for URL building. It generates a URL for a specific function, based on the URL rules defined in the application.

#Here's an example of how to use the url_for() function in Flask:


from flask import Flask, url_for

app = Flask(__name__)

@app.route('/hello/<name>')
def hello(name):
    return f'Hello, {name}!'

with app.test_request_context():
    print(url_for('hello', name='John'))
```
