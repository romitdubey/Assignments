Q1. Explain GET and POST methods.

##GET Method:
The GET method is used to retrieve data from a server. When a GET request is sent, the server responds with the requested data, which can be in the form of HTML, XML, JSON, or other formats. The data is sent as part of the URL query string, and can be viewed in the browser's address bar. GET requests are usually used for requests that are idempotent, meaning that they can be repeated without changing the outcome. For example, when you enter a URL in your browser's address bar and press enter, a GET request is sent to retrieve the web page.

##POST Method:
The POST method is used to submit data to a server. When a POST request is sent, the data is sent in the request body rather than in the URL query string. This makes POST requests more secure than GET requests, because the data is not visible in the URL. POST requests are usually used for non-idempotent requests, meaning that they modify the state of the server. For example, when you fill out a form on a web page and submit it, a POST request is sent to the server with the form data.



Q2. Why is request used in Flask?

In Flask, the request module is used to access incoming request data in the form of HTTP request parameters, form data, and request headers. It is a crucial part of building web applications with Flask because it allows you to read user input and take appropriate actions based on that input.

Q3. Why is redirect() used in Flask?

The redirect() function in Flask is used to redirect the user to a different URL. It is commonly used when you want to redirect the user to a different page after they have completed a certain action, such as submitting a form.

Q4. What are templates in Flask? Why is the render_template() function used?

In Flask, templates are used to generate dynamic HTML pages that can display data and respond to user actions. Templates are written in a markup language such as HTML, and can include placeholders for variables and control structures that are filled in by Flask's template engine.

The render_template() function is used in Flask to render a template and generate an HTML page. It takes the name of the template file as its first argument and any additional variables as keyword arguments. The function looks for the template file in the templates folder in your Flask application directory, and renders it using the template engine.
